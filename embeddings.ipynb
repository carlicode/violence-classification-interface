```python
import ast
```

```python
# Convierte las cadenas en listas de n√∫meros en la columna 'embeddings'
df['embeddings'] = df['embeddings'].apply(lambda x: ast.literal_eval(x))
```

```python
db.add(
    ids=df['ids'].tolist(),
    embeddings=df['embeddings'].tolist(),
    metadatas= df.drop(['ids','embeddings','examples'],axis=1).to_dict('records')
)
```

```python
import chromadb
from chromadb.utils import embedding_functions
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

```python
df = pd.read_csv('embeddings.csv')
```

```python
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
chroma_client = chromadb.Client()
client_persistent = chromadb.PersistentClient(path='/content/data_embeddings')
```

```python
sentence_transformer_ef = embedding_functions.SentenceTransformerEmbeddingFunction(model_name = 'all-MiniLM-L6-v2')
```

```python
db = client_persistent.create_collection(name='violenceb', embedding_function=sentence_transformer_ef)
```

```python
db.add(
    ids=[str(id) for id in df['ids'].tolist()],
    documents=df['examples'].tolist(),
    metadatas= df.drop(['ids','embeddings','examples'],axis=1).to_dict('records')
)
```

```python
db.peek(1)
```

# query

```python
results = db.query(
    query_texts=['mi√©rcoles'],
    n_results=2
)
```
